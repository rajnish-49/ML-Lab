```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

my_list = [1, 2, 3, 4, 5]
my_tuple = (10, 20, 30)
my_set = {100, 200, 300, 200}
my_dict = {'a': 1, 'b': 2, 'c': 3}

print("List:", my_list)
print("Tuple:", my_tuple)
print("Set:", my_set)
print("Dict:", my_dict)

arr = np.array([1, 2, 3, 4, 5])
print("\nNumPy array:", arr)
print("Mean:", arr.mean())

df = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [2, 4, 6, 8, 10]
})
print("\nPandas DataFrame:")
print(df)

slope, intercept, r_value, p_value, std_err = stats.linregress(df['x'], df['y'])
print(f"\nSciPy linear regression slope: {slope}")

plt.figure(figsize=(8, 6))
plt.scatter(df['x'], df['y'], color='blue', label='Data')
plt.plot(df['x'], slope * df['x'] + intercept, color='red', label='Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Matplotlib Plot with SciPy Regression')
plt.legend()
plt.grid(True)
plt.show()
```
